## Variables

A **variable** is a named container for storing a value. You create one using the `=` sign.

```python
name = "Sarah"
city = "London"
age = 28
```

Here we have created three variables. `name` and `city` store text (called **strings**). `age` stores a whole number (called an **integer**).

Run the cell above. Nothing appears — that is normal. Python has stored the values but has not been asked to display them yet.

To see a value, use `print()`.

```python
print(name)
print(city)
print(age)
```

You can also combine values in a print statement:

```python
print("My name is", name, "and I live in", city)
```

---

## Data types

Every value in Python has a **type**. The three most common types you will use are:

| Type | Example | Used for |
|------|---------|----------|
| `str` | `"London"` | Text |
| `int` | `28` | Whole numbers |
| `float` | `12.50` | Decimal numbers |

You can check the type of any variable using `type()`:

```python
print(type(name))
print(type(age))
print(type(12.50))
```

## Reading error messages

Errors are a normal part of coding — every programmer sees them constantly. The important skill is reading them calmly.

Run this cell and look at what appears:

```python
print(Name)
```

You will see a `NameError`. Python is telling you it does not recognise `Name` — because Python is case-sensitive and the variable is called `name`, not `Name`.

Error messages always tell you what went wrong and where. Get into the habit of reading them before asking for help.

## Lists

A **list** is a variable that holds multiple values in order. You create one using square brackets:

```python
items_ordered = ["Beef Burger", "Margherita Pizza", "Caesar Salad"]
```

You can access individual items using their **index** — their position in the list. In Python, counting starts at zero:

```python
print(items_ordered[0])   # Beef Burger
print(items_ordered[1])   # Margherita Pizza
print(items_ordered[2])   # Caesar Salad
```

You can find how many items are in a list using `len()`:

```python
print(len(items_ordered))
```

And you can add an item to the end of a list using `.append()`:

```python
items_ordered.append("Chocolate Brownie")
print(items_ordered)
```
